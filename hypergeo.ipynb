```python
import numpy as np
import math

import matplotlib.pyplot as plt
```

```python
deck_size = 40
hand_size = 5
prob_array = []

for i in range(30):

    starters = i
    non_starters = deck_size - starters

    good_stuff = math.comb(starters, 1)
    bad_stuff = math.comb(non_starters, 4)
    denom = math.comb(deck_size, 5)

    prob = good_stuff*bad_stuff/denom
    prob_array.append(prob)

plt.figure(figsize=(20, 5))
plt.plot(np.arange(0, i+1, 1) , prob_array)
plt.xlabel("x = Number of starters in deck")
plt.xticks(np.arange(0, i, 1), [str(x) for x in [np.arange(0, i, 1)][0]])
plt.ylabel("Probability")
plt.title("Probability of drawing EXACTLY 1 starter with x starters in deck")
plt.grid()
```
